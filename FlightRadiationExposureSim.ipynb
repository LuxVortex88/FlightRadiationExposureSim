```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
earth_radius = 6371  # km
altitudes_km = np.linspace(0, 20, 100) # Flight Altitude from sea level to 20 km
radiation_base = 0.03 # mSv/hr (at sea level)
scale_factor = 2.0 # Increase in radiation dose every 6 km approx

#Simulate radiation exposure by altitude
radiation_dose = radiation_base * np.exp((altitudes_km / 6) * np.log(scale_factor))

# Plot the dose
plt.figure(figsize=(10, 5))
plt.plot(altitudes_km, radiation_dose, color='crimson', linewidth=2)
plt.title("Cosmic Radiation Dose vs Altitude", fontsize=14)
plt.xlabel("Altitude (km)")
plt.ylabel("Radiation Dose (mSv/hr)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
